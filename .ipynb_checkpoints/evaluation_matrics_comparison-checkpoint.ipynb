```python
import matplotlib.pyplot as plt
import cv2
from skimage.metrics import structural_similarity as ssim
```

```python
# Load the images
image1 = cv2.imread('final_project_images/style_images/cathedral.png') # style
image2 = cv2.imread('final_project_images/content_images/miley_cyrus.jpg') # content
image2 = cv2.imread('final_project_gen_images/miley_cyrus_cathedral_baseCNN/3000.png')

# modify images to the same size
image1 = cv2.resize(image1, (image2.shape[1], image2.shape[0]))



# Convert the images to grayscale
image1_gray = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
image2_gray = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

# Compute SSIM between the two images
ssim_index, ssim_map = ssim(image1_gray, image2_gray, full=True)

print(f"SSIM Index: {ssim_index}")

# Display the images and SSIM map
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.title('Content Image')
plt.imshow(image1_gray, cmap='gray')

plt.subplot(1, 3, 2)
plt.title('Generated Image')
plt.imshow(image2_gray, cmap='gray')

plt.subplot(1, 3, 3)
plt.title('SSIM Map')
plt.imshow(ssim_map, cmap='gray')

plt.show()
```
